```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load the complete dataset
housing = fetch_california_housing(as_frame=True)

# Create DataFrame (includes target)
df = housing.frame

# Check the data
print(df.head())
print(df.shape)
print(df.columns)
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["MedHouseVal"], bins=30)

plt.title("Distribution of House Prices")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="MedInc",
    y="MedHouseVal",
    data=df
)

plt.title("Median Income vs House Value")

plt.show()
```

```python
sns.pairplot(
    df[
        [
            "MedInc",
            "HouseAge",
            "AveRooms",
            "MedHouseVal"
        ]
    ]
)

plt.show()
```
